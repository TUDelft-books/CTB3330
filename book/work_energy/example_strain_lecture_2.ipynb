# Evaluate strain energy using SymPy

```python
import sympy as sym
```

```python
L, x, q, GA_C, EI = sym.symbols('L, x, q, GA_C, EI') 
```

```python
 M = -q * (L-x)**2 / 2
E_c_M = sym.integrate(M**2/EI/2,(x,0,L))
display(E_c_M)
```

```python
V = (L-x)*q
E_c_V = sym.integrate(V**2/GA_C/2,(x,0,L))
display(E_c_V)
```
